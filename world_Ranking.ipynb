#### What is the distribution of female-to-male ratio among the universities?

```python
# Convert 'Female to Male Ratio' column to strings
df['Female to Male Ratio'] = df['Female to Male Ratio'].astype(str)

# Plot a histogram for the female-to-male ratio
import matplotlib.pyplot as plt
plt.figure(figsize=(25, 10))
plt.hist(df['Female to Male Ratio'], bins=20, edgecolor='black')
plt.xlabel('Female to Male Ratio')
plt.xticks(rotation=90)
plt.ylabel('Frequency')
plt.title('Distribution of Female to Male Ratio in Universities')
plt.show()
```

```python
# Calculate the correlation between teaching and research environment scores
correlation_teaching_research = df['Teaching'].corr(df['Research Environment'])
correlation_teaching_research
```

#### What is the distribution of overall scores among universities?

```python
# Plot a histogram for the distribution of overall scores
plt.hist(df['Overall Score'].astype(float), bins=20, edgecolor='black')
plt.xlabel('Overall Score')
plt.ylabel('Frequency')
plt.title('Distribution of Overall Scores in Universities')
plt.show()
```

#### How does the student-to-staff ratio vary across different countries?

```python
# Plot a bar chart for student-to-staff ratio by country
plt.figure(figsize=(20, 6))
plt.bar(df['Country'], df['Students to Staff Ratio'].astype(float), color='skyblue')
plt.xlabel('Country')
plt.ylabel('Student to Staff Ratio')
plt.title('Student to Staff Ratio by Country')
plt.xticks(rotation=90)
plt.show()
```

#### Is there any correlation between the size of the student population and the research environment score?

```python
# Scatter plot for student population vs research environment score
plt.scatter(df['Student Population'].astype(float), df['Research Environment'].astype(float), color='green')
plt.xlabel('Student Population')
plt.ylabel('Research Environment Score')
plt.title('Student Population vs Research Environment Score')
plt.show()
```

#### What are the top 10 countries with the highest number of universities represented in the dataset?

```python
# Get the top 10 countries with the highest number of universities
top_countries = df['Country'].value_counts().nlargest(10)

# Plot a bar chart for the top 10 countries
plt.figure(figsize=(12, 6))
top_countries.plot(kind='bar', color='purple')
plt.xlabel('Country')
plt.ylabel('Number of Universities')
plt.title('Top 10 Countries with the Highest Number of Universities')
plt.xticks(rotation=45)
plt.show()
```

#### How do the teaching scores vary among the top 10 universities with the highest overall scores?

```python
# Get the top 10 universities with the highest overall scores
top_universities = df.nlargest(10, 'Overall Score')

# Plot a bar chart for teaching scores in the top 10 universities
plt.figure(figsize=(12, 6))
plt.bar(top_universities['Name'], top_universities['Teaching'].astype(float), color='purple')
plt.xlabel('University')
plt.ylabel('Teaching Score')
plt.title('Teaching Scores in the Top 10 Universities with the Highest Overall Scores')
plt.xticks(rotation=45)
plt.show()
```

#### How does the research environment score compare among the top 5 universities from different countries?

```python
# Get the top 5 universities from different countries
top_5_countries = df.groupby('Country').apply(lambda x: x.nlargest(1, 'Overall Score')).reset_index(drop=True).nlargest(5, 'Overall Score')

# Plot a bar chart for research environment scores in the top 5 universities from different countries
plt.figure(figsize=(12, 6))
plt.bar(top_5_countries['Name'], top_5_countries['Research Environment'].astype(float), color='blue')
plt.xlabel('University')
plt.ylabel('Research Environment Score')
plt.title('Research Environment Scores in the Top 5 Universities from Different Countries')
plt.xticks(rotation=45)
plt.show()
```

#### How does the student population vary among universities in different countries?

```python
# Plot a bar chart for student population by country
plt.figure(figsize=(20, 6))
plt.bar(df['Country'], df['Student Population'].astype(float), color='orange')
plt.xlabel('Country')
plt.ylabel('Student Population')
plt.title('Student Population in Universities by Country')
plt.xticks(rotation=90)
plt.show()
```

#### How does the student-to-staff ratio compare among the top 10 universities with the highest teaching scores?

```python
# Get the top 10 universities with the highest teaching scores
top_teaching_universities = df.nlargest(10, 'Teaching')

# Plot a bar chart for student-to-staff ratio in the top 10 universities with the highest teaching scores
plt.figure(figsize=(12, 6))
plt.bar(top_teaching_universities['Name'], top_teaching_universities['Students to Staff Ratio'].astype(float), color='orange')
plt.xlabel('University')
plt.ylabel('Student to Staff Ratio')
plt.title('Student to Staff Ratio in the Top 10 Universities with the Highest Teaching Scores')
plt.xticks(rotation=45)
plt.show()
```

#### What is the average teaching score for universities in the top 5 countries with the most universities?

```python
# Convert 'Teaching' column to numeric
df['Teaching'] = pd.to_numeric(df['Teaching'], errors='coerce')

# Calculate the average teaching score for universities in the top 5 countries with the most universities
avg_teaching_score_per_country = df.groupby('Country')['Teaching'].mean().nlargest(5)

# Plot a bar chart for average teaching scores in the top 5 countries
plt.figure(figsize=(12, 6))
avg_teaching_score_per_country.plot(kind='bar', color='green')
plt.xlabel('Country')
plt.ylabel('Average Teaching Score')
plt.title('Average Teaching Score in Universities by the Top 5 Countries with the Most Universities')
plt.xticks(rotation=45)
plt.show()
```
